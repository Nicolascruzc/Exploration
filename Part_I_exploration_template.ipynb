```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the plots
sns.set(style="whitegrid")

# Histogram for trip duration (duration_sec)
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="duration_sec", bins=100, kde=True)
plt.title("Distribution of Trip Duration (seconds)")
plt.xlabel("Duration (seconds)")
plt.ylabel("Count")
plt.show()

# Bar chart for user type (user_type)
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="user_type", palette="Set2")
plt.title("Distribution of User Types")
plt.xlabel("User Type")
plt.ylabel("Count")
plt.show()

# Count plot for member gender (member_gender)
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="member_gender", palette="Set1")
plt.title("Distribution of Member Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# Histogram for member age
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="member_birth_year", bins=30, kde=True)
plt.title("Distribution of Member Age")
plt.xlabel("Birth Year")
plt.ylabel("Count")
plt.show()

# Boxplot for trip duration (duration_sec)
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, y="duration_sec", color="skyblue")
plt.title("Boxplot of Trip Duration")
plt.ylabel("Duration (seconds)")
plt.show()

# Boxplot for member age
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, y="member_birth_year", color="salmon")
plt.title("Boxplot of Member Age")
plt.ylabel("Birth Year")
plt.show()

# Bar chart for bike share for all trip
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="bike_share_for_all_trip", palette="Set3")
plt.title("Bike Share for All Trip")
plt.xlabel("Bike Share for All Trip")
plt.ylabel("Count")
plt.show()
```

## Bivariate Exploration

> In this section, investigate relationships between pairs of variables in your data. Make sure the variables that you cover here have been introduced in some fashion in the previous section (univariate exploration).

> **Rubric Tip**: This part (Bivariate Exploration) should include at least one scatter plot, one box plot, and at least one clustered bar chart or heat map.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatterplot
plt.figure(figsize=(10,6))
plt.scatter(data = df, x = 'duration_sec', y = 'member_birth_year')
plt.title('Scatterplot of duration_sec vs member_birth_year')
plt.xlabel('Duration (sec)')
plt.ylabel('member_birth_year')
plt.show()

# Boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data = df, x = 'user_type', y = 'duration_sec')
plt.title('Boxplot of duration_sec for each user_type')
plt.xlabel('User Type')
plt.ylabel('Duration (sec)')
plt.ylim([0, 5000]) # Limiting to get a better visualization
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(data = df, x = 'member_gender', y = 'duration_sec')
plt.title('Boxplot of duration_sec for each member_gender')
plt.xlabel('member_gender')
plt.ylabel('Duration (sec)')
plt.ylim([0, 5000]) # Limiting to get a better visualization
plt.show()

# Clustered Bar Chart
df_grouped = df.groupby(['user_type', 'member_gender']).size().reset_index(name='counts')
pivot_df = df_grouped.pivot(index='user_type', columns='member_gender', values='counts')

pivot_df.plot(kind='bar', stacked=True)
plt.title('Clustered Bar Chart of user_type and member_gender')
plt.xlabel('User Type')
plt.ylabel('Counts')
plt.show()

# Heatmap
# For the heatmap, let's consider the correlation between numerical variables
numerical_vars = ['duration_sec',  'member_birth_year']
corr_matrix = df[numerical_vars].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f")
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

### Talk about some of the relationships you observed in this part of the investigation. How did the feature(s) of interest vary with other features in the dataset?

> **Rubric Tip**: This part (Multivariate Exploration) should include at least one Facet Plot, and one Plot Matrix or Scatterplot with multiple encodings.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Convertir las columnas de tiempo a datetime
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])

# Crear una nueva columna para la hora de inicio
df['start_hour'] = df['start_time'].dt.hour

# FacetGrid
g = sns.FacetGrid(df, col="user_type", height=4, aspect=1.5)
g.map(sns.scatterplot, "start_hour", "duration_sec", alpha=.7)
g.add_legend()
plt.show()

g = sns.FacetGrid(df, col="member_gender", height=4, aspect=1.5)
g.map(sns.scatterplot, "start_hour", "duration_sec", alpha=.7)
g.add_legend()
plt.show()

# PairGrid
g = sns.PairGrid(df, vars=["duration_sec", "start_hour"], hue="user_type")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter, alpha=.7)
g.add_legend()
plt.show()

g = sns.PairGrid(df, vars=["duration_sec", "start_hour"], hue="member_gender")
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter, alpha=.7)
g.add_legend()
plt.show()
```
