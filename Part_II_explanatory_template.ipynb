````python
# load in the dataset into a pandas dataframe

df = pd.read_csv(r'C:\Users\nicol\Downloads\201902-fordgobike-tripdata.csv', encoding="utf-8")
````

## (Visualization 1)

````python
# Set figure size
plt.figure(figsize=(10, 6))

# Create histogram
sns.histplot(data=df, x="member_birth_year", bins=30, kde=True, color='skyblue')

# Add grid
plt.grid(True)

# Calculate mean and median
mean=df['member_birth_year'].mean()
median=df['member_birth_year'].median()

# Add mean line
plt.axvline(mean, color='r', linestyle='--')
plt.text(mean+5,10,f"Mean: {round(mean)}",rotation=90)

# Add median line
plt.axvline(median, color='g', linestyle='-')
plt.text(median-15,10,f"Median: {round(median)}",rotation=90)

# Set title and labels
plt.title("Histogram of Member Birth Year with KDE, Mean and Median")
plt.xlabel("Member Birth Year")
plt.ylabel("Frequency")

# Show plot
plt.show()
````

Here it can be seen that on average the participating people were born in 1985 and the average in 1987, however it is observed that the atypical values ​​are to the left, that is, older people, representing a tail to the left.

## (Visualization 2)

> You should have at least three visualizations in your presentation, but feel free to add more if you'd like!

````python

# Set the overall aesthetic.
sns.set(style="whitegrid")

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(10, 6))

# Create a boxplot
sns.boxplot(data=df, x='user_type', y='duration_sec', ax=ax)

# Add median values on top of the boxplot
medians = df.groupby(['user_type'])['duration_sec'].median().values
nobs = df['user_type'].value_counts().values
nobs = [str(x) for x in nobs.tolist()]
nobs = ["n: " + i for i in nobs]

pos = range(len(nobs))
for tick, label in zip(pos, ax.get_xticklabels()):
    ax.text(pos[tick], medians[tick] + 0.03, nobs[tick],
            horizontalalignment='center', size='x-small', color='w', weight='semibold')

# Set the title and labels
ax.set_title('Boxplot of duration_sec for each user_type', fontsize=15)
ax.set_xlabel('User Type', fontsize=12)
ax.set_ylabel('Duration (sec)', fontsize=12)

# Limit y-axis for better visualization
ax.set_ylim([0, 5000])

# Remove the top and right spines from plot
sns.despine()

# Show the plot
plt.show()
````

In this graph you can see very clearly that subscribers have shorter average times than consumers.

## (Visualization 3)

> You should have at least three visualizations in your presentation, but feel free to add more if you'd like!

````python


# Set the overall aesthetic.
sns.set(style="whitegrid")

# Create a figure and a set of subplots
fig, ax = plt.subplots(figsize=(10, 6))

# Create a scatterplot
sns.scatterplot(data=df, x='duration_sec', y='member_birth_year', ax=ax)

# Add a regression line to the scatterplot
sns.regplot(data=df, x='duration_sec', y='member_birth_year', scatter=False, ax=ax, color='red')

# Set the title and labels
ax.set_title('Scatterplot of duration_sec vs member_birth_year', fontsize=15)
ax.set_xlabel('Duration (sec)', fontsize=12)
ax.set_ylabel('Member Birth Year', fontsize=12)

# Remove the top and right spines from plot
sns.despine()

# Show the plot
plt.show()
````

Here you can see in a very graphic way that in general the times are short since the long ones are seen as atypical data to the right. We also see that there is no relationship between age and times. Observing that even older people have short times.
